```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

# Load the formatted data from CSV
df = pd.read_csv('formatted_sales_data.csv')

# Select features for segmentation
features = ['QUANTITYORDERED', 'PRICEEACH', 'SALES']

# Normalize the features
df_normalized = (df[features] - df[features].mean()) / df[features].std()

# Determine the optimal number of clusters using the elbow method
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42, n_init=10)
    kmeans.fit(df_normalized)
    wcss.append(kmeans.inertia_)
```

```python
# Plot the elbow method
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python

n_clusters = 3

# Apply K-means clustering
kmeans = KMeans(n_clusters=n_clusters, init='k-means++', random_state=42, n_init=10)
df['Cluster'] = kmeans.fit_predict(df_normalized)

# Visualize the clusters using pairplot
sns.pairplot(df, hue='Cluster', palette='tab10', diag_kind='kde')
plt.suptitle('Scatter Plot Matrix with Clusters', y=1.02)
plt.show()

```

```python
# Visualize the clusters using boxplots
plt.figure(figsize=(12, 6))
sns.boxplot(x='Cluster', y='SALES', data=df, palette='tab10')
plt.title('Boxplot of SALES by Cluster')
plt.xlabel('Cluster')
plt.ylabel('SALES')
plt.show()
```

```python

from sklearn.preprocessing import StandardScaler

# Load the formatted data from CSV
df = pd.read_csv('formatted_sales_data.csv')

# Feature Engineering
# 1. Total Sales per Customer
total_sales_per_customer = df.groupby('CUSTOMERNAME')['SALES'].sum().reset_index()
total_sales_per_customer.columns = ['CUSTOMERNAME', 'Total_Sales']

# Merge total sales per customer with the original DataFrame
df = pd.merge(df, total_sales_per_customer, on='CUSTOMERNAME', how='left')

# 2. Number of Orders per Customer
num_orders_per_customer = df['CUSTOMERNAME'].value_counts().reset_index()
num_orders_per_customer.columns = ['CUSTOMERNAME', 'Num_Orders']

# Merge number of orders per customer with the original DataFrame
df = pd.merge(df, num_orders_per_customer, on='CUSTOMERNAME', how='left')

# 3. Average Order Value per Customer
df['Avg_Order_Value'] = df['Total_Sales'] / df['Num_Orders']

# 4. Total Sales per Product
total_sales_per_product = df.groupby('PRODUCTCODE')['SALES'].sum().reset_index()
total_sales_per_product.columns = ['PRODUCTCODE', 'Total_Sales_Product']

# Merge total sales per product with the original DataFrame
df = pd.merge(df, total_sales_per_product, on='PRODUCTCODE', how='left')

# 5. Number of Orders per Product
num_orders_per_product = df['PRODUCTCODE'].value_counts().reset_index()
num_orders_per_product.columns = ['PRODUCTCODE', 'Num_Orders_Product']

# Merge number of orders per product with the original DataFrame
df = pd.merge(df, num_orders_per_product, on='PRODUCTCODE', how='left')

# 6. Average Order Value per Product
df['Avg_Order_Value_Product'] = df['Total_Sales_Product'] / df['Num_Orders_Product']

# Standardize numerical features
numerical_features = ['QUANTITYORDERED', 'PRICEEACH', 'SALES', 'Total_Sales', 'Num_Orders', 'Avg_Order_Value',
                      'Total_Sales_Product', 'Num_Orders_Product', 'Avg_Order_Value_Product']

scaler = StandardScaler()
df[numerical_features] = scaler.fit_transform(df[numerical_features])

# Display the updated DataFrame
print(df.head())
```

```python
df = pd.read_csv('formatted_sales_data.csv')

# Assuming 'ORDERDATE' is the date column, convert it to datetime if it's not already
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'])

# Set 'ORDERDATE' as the index for time series analysis
df.set_index('ORDERDATE', inplace=True)

# Resample the data to a monthly frequency and calculate the total sales for each month
monthly_sales = df['SALES'].resample('M').sum()

# Plot the monthly sales trend
plt.figure(figsize=(12, 6))
monthly_sales.plot()
plt.title('Monthly Sales Trend')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show()
```

```python
# Decompose the time series data to analyze seasonal patterns
from statsmodels.tsa.seasonal import seasonal_decompose

# Decompose the monthly sales data
decomposition = seasonal_decompose(monthly_sales, model='additive')

# Plot the seasonal component
plt.figure(figsize=(12, 6))
decomposition.seasonal.plot()
plt.title('Seasonal Sales Variation')
plt.xlabel('Date')
plt.ylabel('Seasonal Component')
plt.grid(True)
plt.show()
```
